```python
from neo.models.providers.openai import OpenAIResponseModel
from neo.tools.internal_tools.openai import WebSearch
from neo.tools.code_gen import code_execution_tool

system = """You are an expert software engineer. 
Do not write code that interacts with the internet or the operating system. Do not install any packages.
Do not write interactive code.
"""
model = OpenAIResponseModel(model="gpt-5", 
                           instruction=system,
                           timeaware=False,
                           boolean_response=False, 
                           structured_response_model=None,
                           tools=[
                            #    WebSearch, 
                               code_execution_tool
                               ],
                           configs={"reasoning": {"effort": "medium", "summary": "detailed"}},
                           tool_choice="auto",
                           auto_tool_run=False,  # Set to False to return tool calls without executing them
                           )
prompt = """Implement a text-based Tic-Tac-Toe game in Python with the following features: a game board display function, move validation, win
  condition checking, and a simple AI opponent that makes random valid moves. Include a main game loop that handles player input. Use tools"""
test = await model.acreate(prompt, return_response_object=False)  
```
